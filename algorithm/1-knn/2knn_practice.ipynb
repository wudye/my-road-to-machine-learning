```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings('ignore')
```

```python
# read data
pd_data = pd.read_csv('./manualNumber.csv')
pd_data.head()
```

```python
# split data into X and y
X = pd_data.iloc[:, 1:]
y = pd_data.iloc[:, 0]
print(X.shape)
print(y.shape)
print(type(X))
print(X.head(0))
```

```python
# change the data type of X and y to numpy array
X = np.array(X)
y = np.array(y)
print(X[0])
```

Scaler:
MinMaxScaler: (x - min) / (max - min) (clear boundary)
StandardScaler: (x - mean) / std (gaussian distribution)
RobustScaler: (x - median) / IQR (robust to outliers)
Normalizer: x / ||x|| (normalize samples individually)

```python
# data transformation
# transformer = StandardScaler()
transformer = MinMaxScaler()
X = transformer.fit_transform(X)
print(y.shape)
print(X[0])
```

```python
print(y)
```

```python
# split data into trian and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# choose knn classifier set k = 1, use GridSearchCV to find the best k value
knn_model = KNeighborsClassifier(n_neighbors=1)
knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("k = 1 Accuracy: ", accuracy)
param_grid = {'n_neighbors': [3, 5, 7, 9, 11]}
grid_search = GridSearchCV(knn_model, param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best parameters: ", grid_search.best_params_)
print("Best score: ", grid_search.best_score_)
print("Best estimator: ", grid_search.best_estimator_)
print("Best estimator score: ", grid_search.score(X_test, y_test))
```
