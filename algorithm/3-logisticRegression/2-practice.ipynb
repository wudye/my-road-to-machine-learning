from IBM Watson Analytics 的 "Telco Customer Churn" 数据集。
Churn: 客户是否流失
gender: 性别
Partner_att: 是否有伴侣
Dependents_att: 是否有家属
landline: 是否有座机服务
internet_att: 是否有互联网服务
internet_other: 其他互联网服务
StreamingTV: 是否有流媒体电视
StreamingMovies: 是否有流媒体电影
Contract_Month: 是否为月付合同
Contract_1YR: 是否为一年期合同
PaymentBank: 银行支付方式
PaymentCreditcard: 信用卡支付方式
PaymentElectronic: 电子支付方式
MonthlyCharges: 月度费用
TotalCharges: 总费用

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  StandardScaler, MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score
```

```python
df = pd.read_csv('./churn.csv')
df = pd.get_dummies(df)
df.drop(['Churn_No','gender_Male'], axis=1, inplace=True)
df.rename(columns={'Churn_Yes':'flag'},inplace=True)
df.head()
```

```python
x = df[['Contract_Month', 'internet_other', 'PaymentElectronic', 'StreamingTV', 'StreamingMovies']].values
# x = df[[ 'MonthlyCharges', 'TotalCharges']].values
y = df['flag'].values
type (y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
standard_scaler = StandardScaler()
minMax_scaler = MinMaxScaler()
X_train_std = standard_scaler.fit_transform(X_train)
X_test_std = standard_scaler.transform(X_test)
X_train_minmax = minMax_scaler.fit_transform(X_train)
X_test_minmax = minMax_scaler.transform(X_test)
```

```python
lr = LogisticRegression(l1_ratio=1, C=10, solver="liblinear", random_state=42, max_iter=1000)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
lr = LogisticRegression(l1_ratio=1, C=10, solver="liblinear", random_state=42, max_iter=1000)
lr.fit(X_train_minmax, y_train)
y_pred = lr.predict(X_test_minmax)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(f'Precision: {precision_score(y_test, y_pred)}')
print(f'Recall: {recall_score(y_test, y_pred)}')
print(f'F1 Score: {f1_score(y_test, y_pred)}')
print(f'Confusion Matrix: {confusion_matrix(y_test, y_pred)}')
print(f'ROC AUC: {roc_auc_score(y_test, y_pred)}')
print((f'minmax_scaler Accuracy: {accuracy_score(y_test, y_pred)}'))
print((f'minmax_scaler Precision: {precision_score(y_test, y_pred)}'))
print((f'minmax_scaler Recall: {recall_score(y_test, y_pred)}'))
print((f'minmax_scaler F1 Score: {f1_score(y_test, y_pred)}'))
print((f'minmax_scaler Confusion Matrix: {confusion_matrix(y_test,
    y_pred)}'))
print((f'minmax_scaler ROC AUC: {roc_auc_score(y_test, y_pred)}'))
```

```python
lr = LogisticRegression(l1_ratio=0, C=100, solver="liblinear", random_state=42, max_iter=1000)
lr.fit(X_train_std, y_train)
y_pred = lr.predict(X_test_std)
lr = LogisticRegression(l1_ratio=1, C=10, solver="liblinear", random_state=42, max_iter=1000)
lr.fit(X_train_minmax, y_train)
y_pred = lr.predict(X_test_minmax)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
print(f'Precision: {precision_score(y_test, y_pred)}')
print(f'Recall: {recall_score(y_test, y_pred)}')
print(f'F1 Score: {f1_score(y_test, y_pred)}')
print(f'Confusion Matrix: {confusion_matrix(y_test, y_pred)}')
print(f'ROC AUC: {roc_auc_score(y_test, y_pred)}')
print((f'minmax_scaler Accuracy: {accuracy_score(y_test, y_pred)}'))
print((f'minmax_scaler Precision: {precision_score(y_test, y_pred)}'))
print((f'minmax_scaler Recall: {recall_score(y_test, y_pred)}'))
print((f'minmax_scaler F1 Score: {f1_score(y_test, y_pred)}'))
print((f'minmax_scaler Confusion Matrix: {confusion_matrix(y_test,
    y_pred)}'))
print((f'minmax_scaler ROC AUC: {roc_auc_score(y_test, y_pred)}'))
```
