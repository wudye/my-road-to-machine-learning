# task
use l1, l2 and RidgeCV to do the Boston housing price prediction

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge,RidgeCV
from sklearn.metrics import mean_squared_error
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
data_df = pd.read_csv(data_url, sep=r'\s+', skiprows=22, header=None)
data = np.hstack([data_df.values[::2, :], data_df.values[1::2, :2]])
target = data_df.values[1::2, 2]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2, random_state=42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model_lasso = Lasso(alpha=1.0)
model_ridge = Ridge(alpha=1.0)
model_ridge_cv = RidgeCV(alphas=[0.1,  1.0, 2.0, 3.0, 10.0], cv=5)
model_lasso.fit(X_train_scaled, y_train)
model_ridge.fit(X_train_scaled, y_train)
model_ridge_cv.fit(X_train_scaled, y_train)
print(f"Best alpha for RidgeCV: {model_ridge_cv.alpha_}")
```

```python
y_pred_lasso = model_lasso.predict(X_test_scaled)
y_pred_ridge = model_ridge.predict(X_test_scaled)
y_pred_ridge_cv = model_ridge_cv.predict(X_test_scaled)
```

```python
print(f"lasso mse: {mean_squared_error(y_test, y_pred_lasso):.2f}")
print(f"ridge mse: {mean_squared_error(y_test, y_pred_ridge):.2f}")
print(f"ridge cv mse: {mean_squared_error(y_test, y_pred_ridge_cv):.2f}")
```
